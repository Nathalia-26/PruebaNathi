```python
# lectura de tablas tablas
def leer_tabla(tabla, engine):
    query = "SELECT * FROM " + tabla  
    tabla = pd.read_sql(query, engine)
    return tabla
```

```python
# tabla orders 
df_orders=leer_tabla("orders",engine)
df_orders.head()
```

```python
# tabla orderdetails 
df_orderdetails=leer_tabla("orderdetails",engine)
df_orderdetails.head()
```

```python
# tabla customers 
df_customers=leer_tabla("customers",engine)
df_customers.head()
```

```python
# tabla products 
df_products=leer_tabla("products",engine)
df_products.head()
```

```python
# tabla employees 
df_employees=leer_tabla("employees",engine)
df_employees.head()
```

```python
#2 Realiza el cruce entre los DataFrames, asegurÃ¡ndote de utilizar correctamente el parÃ¡metro validate para asegurar la integridad referencial.

# Realizamos cruce entre las tablas orders y orderdetails
merged_df1 = pd.merge(df_orders, df_orderdetails, left_on='orderNumber', right_on='orderNumber', how='inner')
# Realizamos cruce entre merged_df1 y customers
merged_df2 = pd.merge(merged_df1, df_customers, left_on='customerNumber', right_on='customerNumber', how='inner')
# Realizamos cruce entre merged_df2 y products
merged_df3 = pd.merge(merged_df2, df_products, left_on='productCode', right_on='productCode', how='inner')
# Finalmente realizamos cruce merged_df3 y employees
tabla_merged_df = pd.merge(merged_df3, df_employees, left_on='salesRepEmployeeNumber', right_on='employeeNumber', how='inner')
# Mostramos la tabla
tabla_merged_df.head()
```

```python
# 3. Agrega las siguientes columnas, considerando su nombre y la fÃ³rmula asociada
# venta: quantityOrdered*priceEach
# costo: quantityOrdered*buyPrice
# ganancia: considerando las columnas anteriores


# Agragamos la columna ventas
tabla_merged_df['venta'] = tabla_merged_df['quantityOrdered'] * tabla_merged_df['priceEach']
# Agregamos la columna costo
tabla_merged_df['costo'] = tabla_merged_df['quantityOrdered'] * tabla_merged_df['buyPrice']
# Agregamos la columna ganancias
tabla_merged_df['ganancia'] = tabla_merged_df['venta'] - tabla_merged_df['costo']
tabla_merged_df.head()
```

```python
# 4. Â¿CuÃ¡l fue el total de ventas por lÃ­nea de productos? Incluye una fila de totales.

# Agrupamos la lineas de productos y sumamos
ventas_por_linea = tabla_merged_df.groupby('productLine')['venta'].sum().reset_index()
# Calculamos el total general
total = tabla_merged_df['venta'].sum()
# Agregamos la fila 
fila_de_totales = pd.DataFrame({'productLine': ['Total General'], 'venta': [total]})
total_ventas_por_liena = pd.concat([ventas_por_linea, fila_de_totales], ignore_index=True)
# Mostrar el resultado
total_ventas_por_liena.head()
```

```python
#5  Â¿CuÃ¡ntos clientes distintos hicieron compras?

# Cotamos los clientes distintos
distintos_clientes = tabla_merged_df['customerNumber'].nunique()
# Impriminos el resultado
print("Clientes que hicieron distintas compras:", distintos_clientes)
```

```python
# 6. Â¿Existen clientes que aÃºn no han hecho ninguna compra? Â¿CuÃ¡ntos son?

# Lista del total de clientes
total_clientes = df_customers['customerNumber'].tolist()
# Lista de clientes que compraron 
clientes_compradores = tabla_merged_df['customerNumber'].unique()
# Clientes que no han comprado
clientes_no_copradores = set(total_clientes) - set(clientes_compradores)
# Contamos
cantidad_de_clientes_sin_compras = len(clientes_no_copradores)
# Imprimimos el resultado
print("Clientes que no han hecho ninguna compra:", cantidad_de_clientes_sin_compras)
```

```python
# 7. Se solicita la creaciÃ³n de dos reportes, que respondan las preguntas dadas

# a

tabla_merged_df['orderDate'] = pd.to_datetime(tabla_merged_df['orderDate'])
# Filtramos ventas del aÃ±o 2005
ventas_aÃ±o_2005 = tabla_merged_df[(tabla_merged_df['orderDate'].dt.year == 2005)]
# Calcular las ventas 
ventas_por_cliente = ventas_aÃ±o_2005.groupby('customerNumber').agg(
    venta=('venta', 'sum'),
    costo=('costo', 'sum'),
    ganancia=('ganancia', 'sum')
).reset_index()
# Ordenamos de forma decendente para selecionar los primero 10 
top_10_clientes_2005 = ventas_por_cliente.nlargest(10, 'venta')
# DataFrame. mostrar lo 10
top_10_clientes_2005.head(10)
```
